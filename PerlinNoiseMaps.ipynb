# ***Generating Perlin noise map (no octaves)***

```python
import noise
import numpy as np
import matplotlib.pyplot as plt

def generate_perlin_noise(width, height, scale):
    noise_map = np.zeros((height, width))
    for y in range(height):
        for x in range(width):
            noise_value = noise.snoise2(x / scale, y / scale)
            noise_map[y][x] = noise_value
    return noise_map

# Parameters
width = 100
height = 100
scale = 10.0

# Generate Perlin noise
noise_map = generate_perlin_noise(width, height, scale)

# Display noise map
plt.imshow(noise_map, cmap='terrain')
plt.colorbar()
plt.show()
```

# ***Generating Perlin noise map (with octaves)***

```python
import noise
import numpy as np
import matplotlib.pyplot as plt

def generate_perlin_noise(width, height, scale, octaves):
    noise_map = np.zeros((height, width))
    for y in range(height):
        for x in range(width):
            amplitude = 1
            frequency = 1
            noise_value = 0
            for _ in range(octaves):
                noise_value += noise.snoise2(x * frequency / scale, y * frequency / scale) * amplitude
                amplitude *= 0.5
                frequency *= 2
            noise_map[y][x] = noise_value
    return noise_map

# Parameters
width = 100
height = 100
scale = 10.0
octaves = 6

# Generate Perlin noise
noise_map = generate_perlin_noise(width, height, scale, octaves)

# Display noise map
plt.imshow(noise_map, cmap='terrain')
plt.colorbar()
plt.show()
```
